# Cosine similarity 

```python
from sklearn.metrics.pairwise import cosine_similarity
from sentence_transformers import SentenceTransformer
# Load pre-trained SentenceTransformer model
model = SentenceTransformer('all-MiniLM-L6-v2')
# Define reference and predicted sentences
reference = "This school has no training. They put students from all years into one class. All "
translation_gork= "Madrid is the capital of Spain."
translation_gpt = "This school doesn't provide any real education. They lump a handful of students from all years into a single class. What kind of nonsense is this."
# Generate embeddings for the sentences
reference_embedding = model.encode(reference).reshape(1, -1)
prediction_embedding = model.encode(translation_gpt).reshape(1, -1)
# Compute cosine similarity
similarity = cosine_similarity(reference_embedding, prediction_embedding)
print(f"Cosine Similarity: {similarity[0][0]:.4f}")
```

# Sentence similarity

```python
from sentence_transformers import SentenceTransformer, util
# Load pre-trained SentenceTransformer model
model = SentenceTransformer('all-MiniLM-L6-v2')
# Define reference and predicted sentences
reference = "This school has no training. They put students from all years into one class. All "
prediction = "Madrid is the capital of Spain."
# Compute semantic similarity
score = util.cos_sim(model.encode(reference), model.encode(prediction))
print(f"Sentence Similarity Score: {score.item():.4f}")
```

# BERTScore

```python
from bert_score import score
# Define reference and predicted sentences
reference = ["Trường này đào tạo chả có gì. Nhúm mấy sinh viên của tất cả các năm học vào 1 lớp cả năm 1,2,3,4 đều nhốt vào 1 lớp. Dạy tào lao gì không à."]
translation = ["This school provides no education. It locks all students from all years into one class. What kind of nonsense is this."]
# Compute BERTScore
P, R, F1 = score(translation, reference, lang="en", verbose=True)
print(f"BERTScore (Precision): {P[0]:.4f}")
print(f"BERTScore (Recall): {R[0]:.4f}")
print(f"BERTScore (F1): {F1[0]:.4f}")
```

# BLEU Score

```python

from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
# Define reference and predicted sentences
reference = [["The", "capital", "of", "Spain", "is", "Madrid"]]
prediction = ["Madrid", "is", "the", "capital", "of", "Spain"]
# Compute BLEU score
smooth_fn = SmoothingFunction().method4
bleu_score = sentence_bleu(reference, prediction, weights=(0.25, 0.25, 0.25, 0.25), smoothing_function=smooth_fn)
print(f"BLEU Score: {bleu_score:.4f}")
```
